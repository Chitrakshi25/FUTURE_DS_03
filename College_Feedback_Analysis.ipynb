# Actionable Strategies Based on Student Feedback


After analyzing student feedback, here are some strategies to improve the course. The focus is on areas where students gave lower scores.

Overall Approach:
Work on the areas with the lowest ratings and make targeted improvements.

Specific Areas for Improvement

Degree of difficulty of assignments

Strategy: Make assignments more balanced in terms of difficulty.

Actions: Review difficulty vs. learning goals, give examples/resources, allow extra practice, clarify expectations, and support students during Q&A.

Solves doubts willingly

Strategy: Make the doubt-resolution process better.

Actions: Be available for questions, encourage students to ask, use Q&A forums/sessions, and respond quickly.

Course recommendation based on relevance

Strategy: Increase how useful and relevant the course feels.

Actions: Connect lessons to real-world topics or careers, use case studies/examples, discuss value beyond academics, and take student input.

Use of presentations

Strategy: Improve the effectiveness of presentations.

Actions: Try different methods (slides, notes, examples), focus on clear explanations, get feedback from students on how presentations help.
